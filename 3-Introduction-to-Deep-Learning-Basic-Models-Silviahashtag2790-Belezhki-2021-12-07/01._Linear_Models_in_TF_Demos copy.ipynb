```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import tensorflow as tf

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Dense
```

```python
iris_data = pd.read_csv("data/Iris.csv", header  = None) 
iris_data
```

```python
targets = iris_data[4]
```

```python
data = iris_data.drop( 4, axis = 1)
```

```python
data.describe().T
```

```python
data.dtypes
```

```python
targets.unique()
```

```python
targets = targets.replace({"Iris-setosa": 0, "Iris-versicolor": 1, "Iris-virginica": 2})
```

```python
tf.keras.backend.clear_session()
```

```python
log_reg = Sequential([
    Input(shape = (4,)),
    Dense(3, activation = "softmax", kernel_regularizer = "l2")    
])
```

```python
log_reg.summary()
```

```python
log_reg.compile(optimizer = "adam", loss = "sparse_categorical_crossentropy", metrics = ["accuracy"])
```

```python
history = log_reg.fit(data, targets, epochs = 250)
```

```python
history.history
```

```python
plt.plot(history.history["loss"])
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.show()
```

```python
plt.plot(history.history["accuracy"])
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.ylim(0, 1)
plt.show()
```

```python
log_reg.predict(data)
```

```python
tf.keras.metrics.sparse_categorical_accuracy(targets, log_reg.predict(data))
```

```python
tf.keras.metrics.sparse_categorical_accuracy(targets, log_reg.predict(data)).numpy().mean()
```

```python
tf.reduce_mean(tf.keras.metrics.sparse_categorical_accuracy(targets, log_reg.predict(data)))
```

```python
log_reg.evaluate(data, targets)
```
